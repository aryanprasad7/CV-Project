```python
import mrcfile
import numpy as np
from PIL import Image
import os

class ConversionStats:
    def __init__(self):
        self.subfolders_processed = 0
        self.successful_conversions = 0
        self.failed_conversions = 0

def convert_mrc_to_image(mrc_file, output_file):
    try:
        with mrcfile.open(mrc_file, permissive=True) as mrc:
            data = mrc.data

            if data.ndim == 3:
                slice_data = data[data.shape[0]//2, :, :]
            elif data.ndim == 2:
                slice_data = data
            else:
                raise ValueError(f"Unexpected data dimensions: {data.ndim}")

            data_normalized = ((slice_data - slice_data.min()) / (slice_data.max() - slice_data.min()) * 255).astype(np.uint8)
            img = Image.fromarray(data_normalized)

            os.makedirs(os.path.dirname(output_file), exist_ok=True)
            img.save(output_file)
        return True
    except Exception:
        return False

def recursive_convert_mrc_files(input_folder, output_folder):
    stats = ConversionStats()

    for root, dirs, files in os.walk(input_folder):
        mrc_files = [f for f in files if f.endswith('.mrc')]
        if mrc_files:
            stats.subfolders_processed += 1

        for file in mrc_files:
            input_path = os.path.join(root, file)
            relative_path = os.path.relpath(root, input_folder)
            output_path = os.path.join(output_folder, relative_path)
            output_file = os.path.join(output_path, os.path.splitext(file)[0] + '.png')

            if convert_mrc_to_image(input_path, output_file):
                stats.successful_conversions += 1
            else:
                stats.failed_conversions += 1

    print(f"Conversion complete. Statistics:")
    print(f"Subfolders processed: {stats.subfolders_processed}")
    print(f"Successfully converted images: {stats.successful_conversions}")
    print(f"Failed conversions: {stats.failed_conversions}")
```

```python
import os
import mrcfile
import imageio
from tqdm import tqdm

def convert_mrc_to_tiff(input_folder, output_folder):
    """
    Recursively finds all .mrc files in the input folder and converts them to .tiff.
    Creates the equivalent folder structure in the output folder.

    Args:
        input_folder (str): Path to the input folder containing .mrc files.
        output_folder (str): Path to the output folder to save .tiff files.
    """
    # Walk through all files and subdirectories in the input folder
    mrc_files = []
    for root, _, files in os.walk(input_folder):
        for file in files:
            if file.endswith('.mrc'):
                mrc_files.append(os.path.join(root, file))

    # Progress tracking
    with tqdm(total=len(mrc_files), desc="Converting .mrc to .tiff") as pbar:
        for mrc_path in mrc_files:
            # Compute the relative path from the input folder
            relative_path = os.path.relpath(mrc_path, input_folder)
            # Create the equivalent output path
            tiff_path = os.path.join(output_folder, os.path.splitext(relative_path)[0] + '.tiff')
            os.makedirs(os.path.dirname(tiff_path), exist_ok=True)

            # Perform the conversion
            try:
                with mrcfile.open(mrc_path, permissive=True) as mrc:
                    data = mrc.data
                # Save as TIFF
                imageio.imwrite(tiff_path, data[0])  # Export the first slice if it's 3D
            except Exception as e:
                print(f"Error converting {mrc_path}: {e}")

            pbar.update(1)

# Example Usage
convert_mrc_to_tiff("F-actin", "Converted_F-actin")
```

```python
from tqdm import tqdm
import os
from PIL import Image

# Define paths
main_folder = "Converted_CCPs"  # Main directory containing subfolders like Cell_001, Cell_002, etc.
output_folder = "Processed_Dataset"  # Output directory for low_res and high_res
low_res_folder = os.path.join(output_folder, "low_res")
high_res_folder = os.path.join(output_folder, "high_res")

# Create output directories
os.makedirs(low_res_folder, exist_ok=True)
os.makedirs(high_res_folder, exist_ok=True)

# Process each subfolder in the main folder
for subfolder in tqdm(os.listdir(main_folder)):
    subfolder_path = os.path.join(main_folder, subfolder)
    if not os.path.isdir(subfolder_path):
        continue

    # Get all .png files in the current subfolder
    png_files = [f for f in os.listdir(subfolder_path) if f.endswith(".png")]

    # Identify low-res and high-res files
    low_res_files = [f for f in png_files if f.endswith(("01.png", "02.png", "03.png", "04.png", "05.png", "06.png", "07.png", "08.png", "09.png"))]
    high_res_file = "SIM_gt.png"  # High-res file name

    if high_res_file not in png_files:
        print(f"High-resolution file missing in {subfolder}")
        continue

    # Process and resize high-resolution image
    hr_image_path = os.path.join(subfolder_path, high_res_file)
    hr_image = Image.open(hr_image_path).convert("RGB")
    hr_image = hr_image.resize((640, 640))

    # Save the high-resolution image multiple times (once for each low-res image)
    for lr_file in low_res_files:
        hr_output_path = os.path.join(high_res_folder, f"{subfolder}_{lr_file}")
        hr_image.save(hr_output_path)

    # Process and resize low-resolution images
    for lr_file in low_res_files:
        lr_image_path = os.path.join(subfolder_path, lr_file)
        lr_image = Image.open(lr_image_path).convert("RGB")
        lr_image = lr_image.resize((640, 640))

        lr_output_path = os.path.join(low_res_folder, f"{subfolder}_{lr_file}")
        lr_image.save(lr_output_path)

print("Dataset creation completed!")
```

```python
import os
import shutil
from PIL import Image

def resize_image(input_path, output_path, size):
    """Resize an image to the specified size and save it."""
    with Image.open(input_path) as img:
        img = img.resize(size)
        img.save(output_path)

def organize_images_with_duplicates(main_folder, output_folder):
    # Create the output folder structure
    er_folder = output_folder
    high_res_folder = os.path.join(er_folder, 'high_res')
    low_res_folder = os.path.join(er_folder, 'low_res')
    os.makedirs(high_res_folder, exist_ok=True)
    os.makedirs(low_res_folder, exist_ok=True)
    print(er_folder)
    # Iterate over subfolders in the main folder
    for subfolder in os.listdir(main_folder):
        subfolder_path = os.path.join(main_folder, subfolder)
        if not os.path.isdir(subfolder_path):
            continue

        # Identify high-res image (SIM_gt.png)
        # high_res_a_image_path = os.path.join(subfolder_path, "SIM_gt_a.png")
        high_res_image_path = os.path.join(subfolder_path, "SIM_gt.png")
        if not os.path.isfile(high_res_image_path):
            print(f"High-res image not found in {subfolder_path}. Skipping...")
            continue

        # Find all low-res images
        low_res_images = [f for f in os.listdir(subfolder_path) if not f.startswith("SIM_gt") and f.endswith((".png", ".jpg"))]
        # print(high_res_a_image_path)
        # print(high_res_b_image_path)
        # print(low_res_images)
        # break

        for low_res_image in low_res_images:
            # Generate common name
            image_name = f"{subfolder}_{low_res_image}"

            # Copy and resize high-res image
            high_res_dest = os.path.join(high_res_folder, image_name)
            resize_image(high_res_image_path, high_res_dest, (512, 512))

            # Copy and resize low-res image
            low_res_source = os.path.join(subfolder_path, low_res_image)
            low_res_dest = os.path.join(low_res_folder, image_name)
            resize_image(low_res_source, low_res_dest, (64, 64))

# Example usage
main_folder = "Converted_Microtubules"
output_folder = "Microtubules"
organize_images_with_duplicates(main_folder, output_folder)
```
